```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('~/Documents/ml/NA_PDB/prot_nuc.pymolst__diffs-dnar.tsv', sep="\t")
```

```python
data.head()
```

```python
data = data - data.mean()
```

```python
data.head()
```

```python
cov_mat = data.cov()
```

```python
cov_mat
```

```python
eig_values, eig_vectors = np.linalg.eig(cov_mat)
```

```python
e_indices = np.argsort(eig_values)[::-1]
eigenvectors_sorted = eig_vectors[:,e_indices]
eigenvectors_sorted
```

```python
variance_explained = []
for i in eig_values:
    variance_explained.append((i/sum(eig_values))*100)
print(variance_explained)
```

```python
ar=eigenvectors_sorted[:3]
df = pd.DataFrame()
df['1']=ar[0]
df['2']=ar[1]
df['3']=ar[2]
```

```python
import plotly.express as px
fig=px.scatter(df, x='1', y='2')
fig.show()
```
